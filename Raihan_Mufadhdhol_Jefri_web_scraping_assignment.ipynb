Scraping Assignment Using Bs4

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.honda-indonesia.com/news/kesuksesan-honda-hr-v-hantarkan-honda-raih-penghargaan-most-popular-brand-dari-jawa-pos'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
list_content = soup.find('div', id='recent-post-list-sidebar')
```

```python
posts = list_content.find_all('div', class_='clearfix spost')
```

```python
len(posts)
```

```python
posts
```

```python
results = []
for item in posts:
  data = dict()

  # data collection
  data['title'] = item.select_one('.entry-title').text
  data['url'] = item.select_one('a').get('href')
  img_tag = soup.find('img')
  data['img'] = img_tag['src']
  results.append(data)
```

```python
df = pd.DataFrame(results)
df
```

```python
df.to_csv('scraped_data.csv', index=False)
```
